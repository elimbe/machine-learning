```python
import numpy as np
import tensorflow as tf
from keras.models import Sequential
from keras.layers.core import Dense
```

```python
training_data= np.array([[0,0],[0,1],[1,0],[1,1]])
target_data= np.array([[0],[1],[1],[0]])
```

```python
model=Sequential()
```

```python
model.add(Dense(32, input_dim=2, activation='relu'))
```

```python
model.add(Dense(1, activation='sigmoid'))
```

```python
opt_sgd=model.compile(
    loss= 'mean_squared_error',
    optimizer='SGD',
    metrics=['binary_accuracy']
)
```

```python
out_sgd=(model.fit(training_data, target_data, epochs=1000))
```

```python
opt_adam=model.compile(
    loss= 'mean_squared_error',
    optimizer='adam',
    metrics=['binary_accuracy']
)
```

```python
out_adam=model.fit(training_data, target_data, epochs=1000)
```

```python
opt_adadelta=model.compile(
    loss= 'mean_squared_error',
    optimizer='adadelta',
    metrics=['binary_accuracy']
)
```

```python
out_adadelta=model.fit(training_data, target_data, epochs=1000)
```

```python
opt_adagrad=model.compile(
    loss= 'mean_squared_error',
    optimizer='adagrad',
    metrics=['binary_accuracy']
)
```

```python
out_adagrad=model.fit(training_data, target_data, epochs=1000)
```

```python
opt_adamax=model.compile(
    loss= 'mean_squared_error',
    optimizer='adamax',
    metrics=['binary_accuracy']
)
```

```python
out_adamax=model.fit(training_data, target_data, epochs=1000)
```

```python
opt_ftrl=model.compile(
    loss= 'mean_squared_error',
    optimizer='ftrl',
    metrics=['binary_accuracy']
)
```

```python
out_ftrl=model.fit(training_data, target_data, epochs=1000)
```

```python
opt_nadam=model.compile(
    loss= 'mean_squared_error',
    optimizer='nadam',
    metrics=['binary_accuracy']
)
```

```python
out_nadam=model.fit(training_data, target_data, epochs=1000)
```

```python
opt_RMSprop=model.compile(
    loss= 'mean_squared_error',
    optimizer='RMSprop',
    metrics=['binary_accuracy']
)
```

```python
out_RMSprop=model.fit(training_data, target_data, epochs=1000)
```

```python
scores= model.evaluate(training_data, target_data)
```
